```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

width = 4  # Because it looks good
fontsize = 12
rc = {'figure.figsize': (width, width),
      'font.size': fontsize,
      'axes.labelsize': fontsize,
      'axes.titlesize': fontsize,
      'xtick.labelsize': fontsize,
      'ytick.labelsize': fontsize,
      'legend.fontsize': fontsize-2.5}
sns.set(rc=rc)
sns.set_style('ticks')
```

## 1) VibraCLIP Performance without Molecular Mass

```python
ir_dict_no_mass = {'1': 0.12432305156581117, '5': 0.35664390550192293, '10': 0.4806530099678204, '15': 0.5511341338984381, '20': 0.6001883682599481, '25': 0.6358213641001491}
ir_raman_dict_no_mass = {'1': 0.5512911074483949, '5': 0.8394160583941606, '10': 0.8959265363786202, '15': 0.9190016482222746, '20': 0.9318734793187348, '25': 0.9403500510164038}
ir_raman_dict_2_no_mass = {'1': 0.6292284750019622, '5': 0.8563692017894985, '10': 0.9041676477513539, '15': 0.9243387489208068, '20': 0.9351699238678283, '25': 0.9430186013656698}
```
